```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = {20: 0.5, 35: 0.7, 50: 0.0}
B = {20: 0.8, 35: 0.2, 50: 1.0}
```

```python
def union(A, B):
    return {x: max(A.get(x, 0), B.get(x, 0)) for x in set(A) | set(B)}


def intersection(A, B):
    return {x: min(A.get(x, 0), B.get(x, 0)) for x in set(A) | set(B)}


def complement(A):
    return {x: 1 - A[x] for x in A}
def plot_set(set, title):
    plt.figure()
    plt.plot(list(set.keys()), list(set.values()), marker='o')
    plt.title(title)
    plt.xlabel('Age')
    plt.ylabel('Membership Value')
    plt.grid(True)
    plt.show()
```

  2) Write a Python program to demonstrate the following properties on Fuzzy sets

    a.De Morgan’s Law
    b.Transitivity
    c.Distributivity

```python
def de_morgan_law(A, B):
    not_A_union_B = complement(union(A, B))
    not_A_intersection_not_B = intersection(complement(A), complement(B))
    return not_A_union_B, not_A_intersection_not_B
def transitivity(A, B, C):
    return intersection(A, intersection(B, C))


def distributivity(A, B, C):
    return union(A, intersection(B, C))
C = {20: 0.3, 35: 0.5, 50: 0.1}
print(de_morgan_law(A, B))
print(transitivity(A, B, C))
print(distributivity(A, B, C))
```

3) Write a Python program to plot fuzzy membership function graphs using
the following membership functions:


    a.Trapezoidal 
    b.Triangular 
    c.Gaussian 

```python

def plot_membership_function(x, y, title):
    plt.figure()
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('Membership Value')
    plt.grid(True)
    plt.show()


def trapezoidal_mf(x, a, b, c, d):
    return np.maximum(0, np.minimum((x-a)/(b-a), np.minimum(1, (d-x)/(d-c))))


def triangular_mf(x, a, b, c):
    return np.maximum(0, np.minimum((x-a)/(b-a), (c-x)/(c-b)))


def gaussian_mf(x, c, sigma):
    return np.exp(-0.5 * ((x - c) / sigma) ** 2)


x = np.linspace(0, 100, 500)
plot_membership_function(x, trapezoidal_mf(x, 20, 30, 40, 50), 'Trapezoidal Membership Function')
plot_membership_function(x, triangular_mf(x, 20, 35, 50), 'Triangular Membership Function')
plot_membership_function(x, gaussian_mf(x, 35, 10), 'Gaussian Membership Function')
```
